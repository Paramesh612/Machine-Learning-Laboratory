```python
# AND gate 
# Identify weights -> [input] -> Summation -> Output = Activation(summation)  
```

```python
import math
import matplotlib.pyplot as plt

# Q1
def and_prediction(x1, x2):
    sum_val = summation(x1,x2)
    predicted_output = activation(sum_val)
    return predicted_output

def summation(x1,x2):
    return (2*x1)+(2*x2)-3

def activation(sum_val): # Sigmoid
    val = 1/(1+math.e**(-sum_val))
    if(val<=0.5):
        return 0
    else :
        return 1
```

```python
print("Prediction:","\n0 , 0 = ",and_prediction(0,0),"\n0 , 1 = ",and_prediction(0,1),"\n1 , 0 = ",and_prediction(1,0),"\n1 , 1 = ",and_prediction(1,1))
```
