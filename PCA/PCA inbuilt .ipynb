```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python

# Sample 2D data
X = np.array([[2.5, 2.4],
              [0.5, 0.7],
              [2.2, 2.9],
              [1.9, 2.2],
              [3.1, 3.0],
              [2.3, 2.7],
              [2, 1.6],
              [1, 1.1],
              [1.5, 1.6],
              [1.1, 0.9]])
```

```python

# Apply PCA
pca = PCA(n_components=1)  # Reduce to 1 dimension
X_pca = pca.fit_transform(X)
X_pca
```

```python

# Inverse transform to project back to 2D for visualization
X_projected = pca.inverse_transform(X_pca)

# Plot original vs projected
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], color='blue', label='Original')
plt.scatter(X_projected[:, 0], X_projected[:, 1], color='red', label='PCA Projection')
plt.legend()
plt.title("PCA Projection (1D)")
plt.grid(True)
plt.show()
```
