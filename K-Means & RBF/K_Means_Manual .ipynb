```python
c1 = x[0]
c2 = x[1]
cluster = np.zeros(len(x))
cluster
```

```python
def EuclideanDistance(x,y):
    return np.sqrt((x[0]-y[0])**2 + (x[1]-y[1])**2)
```

```python
epoch = 10000
```

```python
for _ in range(epoch):
    for i in range(len(x)):
        cluster[i] = 1 if EuclideanDistance(x[i],c1)<EuclideanDistance(x[i],c2) else 2

    #update c1:
    new_c=np.array([0.0,0.0])
    count = 0
    for i in range(len(cluster)):
        if(cluster[i]==1):
            new_c+=x[i]
            count+=1
    if(count>0): c1=new_c/count


    #update c2:
    new_c=np.array([0.0,0.0])
    count = 0
    for i in range(len(cluster)):
        if(cluster[i]==2):
            new_c+=x[i]
            count+=1
    if(count>0): c2=new_c/count
```
