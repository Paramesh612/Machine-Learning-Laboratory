```python
from sklearn.svm import SVC
from sklearn.datasets import make_circles
from sklearn.metrics import accuracy_score
```

```python
model = SVC(kernel='rbf')
```

```python
x, y = make_circles(n_samples=1000, factor=0.5, noise=0.1, random_state=42)
```

```python
x
```

```python
model.fit(x,y)
```

```python
pred=model.predict(x)
```

```python
model.support_vectors_
```

```python
accuracy_score(pred,y)
```

```python
plt.scatter(x[:,0],x[:,1],c=pred,cmap='viridis')
# plt.scatter(model.support_vectors_[:,0],model.support_vectors_[:,0],c='r')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_max , x_min = x[:,0].max()+1 , x[:,0].min()-1
y_max , y_min = x[:,1].max()+1 , x[:,1].min()-1
xx , yy = np.meshgrid(np.arange(x_min,x_max,0.01),np.arange(y_min,y_max,0.01))
z = model.predict(np.c_[xx.ravel(),yy.ravel()])
z = z.reshape(xx.shape)

plt.scatter(x[:,0],x[:,1],c=pred,cmap='viridis')
plt.contourf(xx,yy,z,cmap='bwr',alpha=00.1)
```
