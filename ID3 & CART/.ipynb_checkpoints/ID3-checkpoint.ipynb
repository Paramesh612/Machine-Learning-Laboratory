```python
import pandas as pd
import math

# Q1 ID3


# Tree

class TreeNode :
    def __init__(self,el):
        self.data = el
        self.children = {}
        
    def add_child(self, key , node):
        self.children[key] = node

def pre_order_traversal(node):
    if node is None:
        return
    print("val: ",node.data)
    print("Children",node.children.keys())
    print()

    for child in node.children.values():
        pre_order_traversal(child)
        

age = TreeNode("age")
student = TreeNode("student")
credit_rating = TreeNode("credit_rating")
yes = TreeNode("yes")

age.add_child("youth",student)
age.add_child("mid_age",yes)
age.add_child("old",credit_rating)

pre_order_traversal(age)
```
